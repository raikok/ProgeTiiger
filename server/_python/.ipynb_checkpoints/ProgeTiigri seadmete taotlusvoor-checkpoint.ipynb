```python
import pandas as pd
import re
```

```python
filename = "PT seadmete taotlusvoor 2014-2021.xlsx"
```

```python
taotlusvoor = pd.read_excel(filename, header=4)
```

```python
devicesAmounts = {}

def sumInDict(key, value):
    if key in devicesAmounts:
        devicesAmounts[key] += value
        return
    devicesAmounts[key] = value
    
firstRegex = re.compile('([\n]+|(;\s)+|[^põhikompl](\.\s)+|(,\s)+)')
secondRegex = re.compile('^\d(\d| )[^ MB][^esi]')

lastKey = 'NaN'
parsedDevices = 0
unparsedDevices = 0

for row in taotlusvoor.itertuples():
    if row[5] == row[5]: #check for !NaN
        allDevices = firstRegex.split(row[5]) # Split on [ . , ; \n ]
        
        for device in allDevices:
            if device:
                deviceSplitOnBracket = device.split('(') # Split on all brackets
                bracketSplits = len(deviceSplitOnBracket) - 1 # Get last bracket start (usually before 'tk')
                amount = 1
                if bracketSplits > 0:
                    # Split on 'tk' in last bracket
                    amountsSplit = deviceSplitOnBracket[bracketSplits].split('tk')
                    if len(amountsSplit) > 1:
                        # Get int before 'tk'.
                        try:
                            strippedDevice = deviceSplitOnBracket[0].strip()
                            amount = amountsSplit[len(amountsSplit) - 2].strip()

                            if secondRegex.match(strippedDevice): # if device had '.' in it's name
                                if lastKey != 'NaN':
                                    strippedDevice = lastKey + " - " + strippedDevice
                                    
                                    #print('addedWithLast')
                                    #print(strippedDevice)
                                    #print('addedWIthLast\n---')

                                    sumInDict(strippedDevice, int(amount))
                                    devicesAmounts.pop(lastKey, "")
                                else:
                                    sumInDict(strippedDevice, int(amount))
                            else:
                                sumInDict(strippedDevice, int(amount))
                            parsedDevices += 1
                            lastKey = strippedDevice
                        except:
                            unparsedDevices += 1
                            continue
                            #print('exceptionOnParse')
                            #print(deviceSplitOnBracket[0].strip())
                            #print(amount)
                            #print('exceptionOnParse\n---')
                            
                else:
                    if re.match('^(LME|EV3)$', device):
                        parsedDevices += 1
                        sumInDict(device.strip(), 1)
                        continue
                    if not (re.search('^(;|,|\.|\)\.)', device)):
                        unparsedDevices += 1
                        continue
                        #print(device)
                        #print('---')
                    else:
                        #print(device)
                        continue
                    
print(str(parsedDevices) + " devices were successfully parsed")
print(str(unparsedDevices) + " devices failed to parse")
#print({k: v for k, v in sorted(devicesAmounts.items(), key=lambda item: item[1])})
```

```python
# optimized

devicesAmounts = {}

firstRegex = re.compile('([\n]+|(;\s)+|[^a-z](\.\s)+|(,\s)+)')
deviceAmountRegex = re.compile('(\d+(\s)?)tk')
cleanDeviceAmountRegex = re.compile('(\s(\()?\d+(\s)?)tk')

def sumInDict(key, value):
    key = cleanDeviceAmount(key)
    value = int(str(value).strip())
    if len(key) > 0 and isinstance(key, str) and isinstance(value, int):
        if key in devicesAmounts:
            devicesAmounts[key] += value
            return
        devicesAmounts[key] = value

    
def cleanDeviceAmount(input):
    cleaned = cleanDeviceAmountRegex.split(input)[0]
    if not re.search('(tk)+', cleaned):
        return cleaned
    else:
        return ""


lastKey = 'NaN'
lastHadAmount = True

for row in taotlusvoor.itertuples():
    if row[5] == row[5]: #check for !NaN
        allDevices = firstRegex.split(row[5]) # Split on [ . , ; \n ]
        i = 0
        
        for device in allDevices:
            if device and not (re.search('^(;|,|\.|\)\.)', device)):
                
                deviceAmountWithSuffix = deviceAmountRegex.search(device)
                if (deviceAmountWithSuffix): # device had amount with it
                    if lastHadAmount: #adding amount to total
                        sumInDict(cleanDeviceAmount(device), deviceAmountWithSuffix.group(1).strip())
                    else:  #if it found number but last one didn't have one
                        
                        if len(cleanDeviceAmount(device)) < 2:
                            # if this amount doesn't have device in front of it, then the amount must be the last one's.
                            sumInDict(lastKey, deviceAmountWithSuffix.group(1))
                        
                        else: #this number has device infront of it. Therefore last one was a single device. And this amount
                            # belongs to current device
                            sumInDict(lastKey, 1)
                            sumInDict(device, deviceAmountWithSuffix.group(1))
                        lastKey = 'NaN'
                        lastHadAmount = True
                    
                else: # if device had no amount with it, then it's amount must come later or is just a single device
                    if not lastHadAmount: # last device didn't have an amount either
                        # then last one was single
                        sumInDict(lastKey, 1)
                        lastKey = cleanDeviceAmount(device)
                    else: # last device had an amount
                        # is it the last device in this row?
                        if i + 1 >= len(allDevices):
                            sumInDict(device, 1) #can't be any after this, must be single
                        else:
                            lastHadAmount = False #it's amount will come later
                            lastKey = cleanDeviceAmount(device)
                i+=1

# optimized
```

```python
import copy

def place_item_total_to_tag(item, value, tag_obj, found_match):
    if tag_obj is not str:
        tag_list = list(tag_obj.keys())
        for tag in tag_list:
            if tag != "total":
                if (re.search(("(?i)" + str(tag)), item)):
                    if "total" in tag_obj[tag]:
                        tag_obj[tag]["total"] += value
                    found_match = place_item_total_to_tag(item, value, tag_obj[tag], True)
    return found_match

copy_dict = copy.deepcopy(devicesAmounts)
tags_copy = copy.deepcopy(tags)

def from_dict_to_tags():
    for item in list(copy_dict.keys()):
        if item in copy_dict:
            match = place_item_total_to_tag(item, copy_dict[item], tags_copy, False)
            if match:
                copy_dict.pop(item)

#print(place_item_total_to_tag(item, 10, tags, False))
from_dict_to_tags()
#print(copy_dict)
print(tags_copy)
```
